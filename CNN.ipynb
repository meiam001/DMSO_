```python
import numpy as np
import pandas as pd
import tensorflow.keras
from tensorflow.keras.applications.vgg19 import VGG19
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Dropout, Flatten
from PIL import Image
from scipy.ndimage import imread
import os
from tensorflow.python.keras.layers import Conv2D
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from collections import defaultdict
from IPython.display import Image
import matplotlib.pyplot as plt
import pickle
from skimage.transform import resize
import random
```

```python
path = '.'
flowers = os.path.join(path, 'flowers-recognition\\flowers')
dandelions = os.path.join(flowers, 'dandelion')
```

```python
# reading flowers into memory

pics = {}
poopoo_memory = True


# imread = plt.imread
for flower_type in os.listdir(flowers):
    pics[flower_type] = []
    flower_dir = os.path.join(flowers,flower_type)
    flower_list = os.listdir(flower_dir)
    for c, flower in enumerate(flower_list):
        if poopoo_memory:
            if c == 250:
                break
        flower = os.path.join(flower_dir, flower)
        try:
            pics[flower_type].append(resize(imread(flower),(200,200,3)))
        except:
            print('{} aint no image foo!'.format(flower))
```

```python
fig=plt.figure(figsize=(200, 200))
images_to_show = 9
for item in range(1,images_to_show+1):
    rando_rose = pics['rose'][random.randint(0,250)]
    fig.add_subplot(images_to_show/3,3,item)
    plt.imshow(rando_rose)
plt.show()
# pics.keys()
```

```python
bm = VGG19(weights=None)
bm.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

[tensorflow.keras.callbacks.EarlyStopping(monitor='val_acc', patience=3, verbose=1)]
bm.summary()

for object in bm.layers:
    print(object.trainable)
```
